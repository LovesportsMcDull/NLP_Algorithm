```python
import pandas as pd
import numpy as np
```

```python
daily_data = pd.read_csv('training_dataset/daily_dataset.csv', index_col= 0)
hourly_data = pd.read_csv('training_dataset/hourly_dataset.csv', index_col= 0)
per5min_data = pd.read_csv('training_dataset/per5min_dataset.csv', index_col=0)

del hourly_data['train or test']

daily_data.index = pd.to_datetime(daily_data.index)
hourly_data.index = pd.to_datetime(hourly_data.index)
per5min_data.index = pd.to_datetime(per5min_data.index)
```

```python
result = pd.read_csv('sample_submission.csv', index_col=0)
result.index = pd.to_datetime(result.index)
result = result.astype('float64')
```

```python
period1 = pd.Timedelta(days = 7)
period2 = pd.Timedelta(hours = 1)
```

```python
for date in result.index:
    print(date)
    for col in range(20):
        period1 = pd.Timedelta(days = 7)
        while (hourly_data.loc[date - period1][col] < 0) or pd.isna(hourly_data.loc[date - period1][col]) or pd.isna(hourly_data.loc[date - period1 + period2][col]) or pd.isna(hourly_data.loc[date - period1 - period2][col]):
            period1 = period1 + pd.Timedelta(days = 7)
        result.loc[date][col] = hourly_data.loc[date - period1][col]
```

```python
result.to_csv('sample_submission.csv')
```
